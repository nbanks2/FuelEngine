# Project Overview

Our project seeks to build a model that calculates the power of a coupe model car’s fuel engine based upon the properties of the engine. We will consider how the cylinders, valves per cylinder, displacement, bore x stroke, compression ratio, max torque, max torque @rpm, max power @rpm, fuel delivery, engine type, turbo, catalyst, and fuel tank impact the maximum power capacity of the engine. Essentially, we will determine which factors contribute most significantly to a coupe car’s engine power capacity. 

We will build this model by scraping data from http://www.car-data.com on coupe model cars, and performing linear regression analyses for each individual attribute of the fuel engine. Each attribute will be considered the dependent variable (y axis) in our analyses, and the maximum capacity of the fuel engine will be the independent variable (x axis). 

A positive correlation coefficient indicates that the dependent variable likely has a positive contribution to the maximum power capacity of the fuel engine, while a negative correlation coefficient indicates the the dependent variable likely has a negative contribution to the maximum power capacity of the fuel engine. The closer the correlation coefficient is to zero, the less contribution that particular variable makes. 

We hypothesize that displacement and turbo will have the biggest positive impact on the maximum power capacity of the engine and fuel delivery will have the biggest negative impact on the maximum power capacity of the engine.

## Data Sources and Rationale


For our first iteration, we obtained our data without writing any code. Rather, we manually input the data from https://www.car-data.com into an excel spreadsheet. We decided to use this website because it contains the exact set of data we need in a comprehensible format.  The data we obtained from https://www.car-data.com is going to help us construct our predictive model by providing us with the fuel engine statistics that we require to procure thorough, accurate conclusions. It is important to note that for the purposes of our research we are considering turbos and superchargers to be the same (they both serve to push air into the engine at an accelerated rate), despite major mechanical differences.
We chose to use coupe model cars because we wanted a standard model of cars to compare engines, and coupe model cars were likely to have a wider variety of engines than some other models of cars.  We used 100 samples of different cars to build a wide ranging list of engines.
	


## Data Cleaning Plan

Originally, the data was in a format on the website that was impossible to be exported in a way that could be cleaned.  In order to harvest the data, we created an excel spreadsheet and copied the whole set of values from the website. By doing this, we ended up with columns of data that corresponded to each individual car. Each column had a format where every odd numbered row was the question about the car and every even numbered row was the answer to that question.
	
Our data cleaning plan is to convert every answer into a single value that is appropriate for the question being asked (if it’s a number followed by units and/or text, reduce it to only being a number).  Since the questions are all repeated in the same order, we can build a list of values for every question.
